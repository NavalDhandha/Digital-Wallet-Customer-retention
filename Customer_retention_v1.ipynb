# Cohort Analysis

## Customer retention by month

```python
df_trans['cohort_month'] = df_trans.groupby('user_id')['transaction_month'].transform('min')

def get_month_diff(d1,d2):
    return (d2.year - d1.year) * 12 + (d2.month - d1.month)
df_trans['cohort_index'] = df_trans.apply(lambda row: get_month_diff(row['cohort_month'],row['transaction_date']),axis= 1)

cohort_counts = df_trans.groupby(['cohort_month','cohort_index'])['user_id'].nunique().reset_index()
cohort_pivot = cohort_counts.pivot(index = 'cohort_month', columns = 'cohort_index', values = 'user_id')

plt.figure(figsize = (12,6))
sns.heatmap(cohort_pivot, annot = True, fmt='g',cmap = 'Oranges')
plt.title('Monthly customer count')
plt.xlabel('Cohort_index(Month since first transaction)')
plt.show()
```

```python
cohort_sizes = cohort_pivot.iloc[:, 0]
retention = cohort_pivot.divide(cohort_sizes, axis=0).round(3)

plt.figure(figsize = (12,6))
sns.heatmap(retention, annot = True, fmt='.0%',cmap = 'Oranges')
plt.title('Retention rate')
plt.xlabel('Cohort_index(Month since first transaction)')
plt.show()
```

Change in the monthly customer count shown in the heatmap suggest inconsistent customer acquisition throughout the period. Most recent cohorts are smaller, shows reduced acquisition. This could attribute to few causes such as lower marketing spend, rising competition, or quality.

Monthly retention rate also suggests significant drop in retention after first month despite high transaction amount in recurring bills. All cohort could only sustain customers in single digit, which means most users only try it once and then did not return. Another trend suggest there is better retention for first few month which could mean poor customer experience, inconvenience and disruptions while using the service.


# Transaction amount retention by month

```python
avg_amount = df_trans.groupby(['cohort_month','cohort_index'])['product_amount'].mean().reset_index()
avg_amount_pivot = avg_amount.pivot(index = 'cohort_month', columns = 'cohort_index', values = 'product_amount')

plt.figure(figsize = (12,6))
sns.heatmap(avg_amount_pivot, annot = True, fmt='.0f',cmap = 'Oranges')
plt.title('Avg transaction amount')
plt.xlabel('Cohort_index(Month since first transaction)')
plt.show()
```

In general it can be seen that customer intial spending is relatively high between 4800 to 5200 in month 0. Later average amount does varies but it can be seen that the average amount still satys between 4000 to 5000. Average amount fluctuates between months and cohorts. In some cases, it does increases slightly for small retained users.

This trend suggests that even though customer retantion is low but the retained users continues to be higher value who use digital wallet.

# Category specific retantion rate

```python
cohort_category = df_trans.groupby(['cohort_month','cohort_index','product_category'])['user_id'].nunique().reset_index()
cohort_category_pivot = cohort_category.pivot(index = ['product_category','cohort_month'], columns = 'cohort_index', values = 'user_id')


category_retention = cohort_category_pivot.divide(cohort_category_pivot.iloc[:, 0], axis=0).round(3)

top_category = df_trans.groupby('product_category')['product_amount'].sum().sort_values(ascending = False).head(5)
filtered_retention = category_retention.loc[top_category.index]

plt.figure(figsize = (12,12))
sns.heatmap(filtered_retention, annot = True, fmt='.0%',cmap = 'Oranges')
plt.title('Retention rate for product category')
plt.xlabel('Cohort_index(Month since first transaction)')
plt.show()
```

Similar to overall retation rate, retantion rate by product category also drops sharply after month 0. Aditionally this analysis confirms that higher value customers are more loyal as retantion rate for these cateogries are generally higher than total retantion. i.e. Users returns to make more transaction. 
Few categories show better retantion after month 1 indicating potential for builing trust and loyalty in the product.

Recommendation:

This cohort analysis of digital wallet shows serious problem of retantion and loyalty. 
With the further analyses on amount and product categories more insights can be collected and some recommendation can be made.

Reason behind this can post on borading experience.  To understand this funnel analysis to identify drop off points as well as session recording and in app analytics to identify painpoints. Then, those analytics can be used to optimize post onboarding experience (e.g rearrange home page, better walk-through of the app. Use A/B trsting methods to identify customers' preference on app navigation and journey.

2. Focusing on retaining higher value customer and better performing categories : From the analysis on average amount and categories it is visible that higher value customer tend to be those making higher recurring(possibly) payments (bills, subscription and essential services). Segment users based on initial month(month 0) spending and target custom campaigns to retain those customers at early stage. Promote and nudge customers through in-app banners and push notification to promote recurring payments. Automatic monthly payments or calendar notification can be the way to go.

3. Launch early campaigns: From the analysis it can be seen that most of the users drop from month 0 to month 1 this gives short window for retantion. Event based engagement is necassary. Common tactic is to try and customers in first 7 to 15 days through events triggers and customized campaigns. Also targeted offers can be pushed to users around 20-25 days to retain customers before they drop.
